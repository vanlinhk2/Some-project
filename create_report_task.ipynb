```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path = r"D:/1_learn_code/code_ds/Getting_data/other_learn/"
df = pd.read_csv("./sales2019_1.csv")
```

```python
df.head()
```

```python
frames = []
all_length = []
for file in os.listdir(path):
    if file.endswith(".csv"):
        filepath = path + file
        df0 = pd.read_csv(filepath)
        frames.append(df0)
        result = pd.concat(frames)
        length_1month = len(df0.index)
        all_length.append(length_1month)
df = result
df
```

```python
print(sum(all_length))
```

```python
df["month"] = df['Order Date'].str.slice(start = 0, stop = 2, step = 1)
df
```

# Get rid of NaN and Or value

```python
df = df.dropna(how = "all")
df
```

# Task 3: Reporting

```python
df["Price Each"] = pd.to_numeric(df["Price Each"], downcast = 'float')
df["Quantity Ordered"] = pd.to_numeric(df["Quantity Ordered"], downcast = 'integer')
```

```python
df["Sales"] = df["Quantity Ordered"] * df["Price Each"]
df
```

```python
moving_col = df.pop('Sales')
df.insert(4, "Sales", moving_col)
df
```

```python
sale_values = df.groupby("month").sum()["Sales"]
sale_values.max()
```

```python
months = range(1, 7)
plt.bar(x=months, height=sale_values)
plt.xticks(months)
plt.xlabel("Month")
plt.ylabel("Sale in U5D")
```
