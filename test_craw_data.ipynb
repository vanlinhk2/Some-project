```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.formula1.com/en/results.html/2023/drivers.html'
```

```python
response = requests.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser') 
```

```python
soup.title.text
```

```python
soup.div
```

```python
table = soup.find('table', class_ ='resultsarchive-table')
```

```python
table
```

```python
rows = table.find_all('tr')
```

```python
rows[0]
```

```python
rows[1]
```

```python
poisi, name,  national, car_name, pts = [], [], [], [], []
for row_id, row in enumerate(rows):
    columns = row.find_all('td')
    #print(f'row = {row_id}: {columns}')
    if columns:
        position = columns[1].text.strip()
        poisi.append(position)
        driver_name = columns[2].text.strip().split('\n')
        driver_name = ' '.join(driver_name[:-1])
        name.append(driver_name)
        nationality = columns[3].text.strip()
        national.append(nationality)
        car = columns[4].text.strip()
        car_name.append(car)
        points = columns[5].text.strip()
        pts.append(points)
        print(f"Poisition : {position}\
                \nName :{driver_name}\
                \nNationality: {nationality}\
                \nCar :{car}\
                \nPoints :{points}\n")
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(zip(poisi, name,  national, car_name, pts)), columns = ['Poisition', 'Driver name', 'Nationality', 'Car', 'Points'])
```
