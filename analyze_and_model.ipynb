```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('dulieu.xlsx')
df.head(3)
```

```python
df[df.columns[4]].unique()
```

```python
df['Group'] = df['Group'].str.strip()
```

```python
df['Group'].unique()
```

```python
df['Giới tính'].unique()
```

```python
indices_to_swap = df[df['Group'].str.len() > 1].index
indices_to_swap
```

```python
df.loc[indices_to_swap]
```

```python
tmp = df.loc[indices_to_swap]
```

```python
tmp[['Group', 'Giới tính']] = tmp[['Giới tính', 'Group']]
```

```python
df.update(tmp)
```

```python
df['Group'].unique()
```

```python
df['Giới tính'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
data = df.copy()
label_encoder = LabelEncoder()
```

```python
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = label_encoder.fit_transform(data[column])
```

```python
data.head()
```

```python
data['Ngành học'].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

X = data.drop(columns=['Ngành học'])
y = data['Ngành học']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

### Random Forest

```python
# Xây dựng mô hình Random Forest
rf_model = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf = confusion_matrix(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print(f'Confusion matrix: \n{conf}')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [80, 100, 130, 150,180, 210,220,230,300],
    'max_depth': [5, 8,10, 13, 17, 22],
    'min_samples_split': [2,3,4, 5],
    'min_samples_leaf': [1, 2,3,4]
}
```

```python
rf_model1 = RandomForestClassifier(random_state=42)
grid = GridSearchCV(estimator= rf_model1, param_grid= param_grid, cv = 5, scoring='accuracy', n_jobs = -1)
```

```python
from datetime import datetime
```

```python
print(datetime.now())
grid.fit(X,y)
print(datetime.now())
```

```python
print("Best parameter:", grid.best_params_)
print("Best score: ", grid.best_score_)
```

### Decision tree

```python
# mô hình decision tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

```python
dt_pred = dt_model.predict(X_test)

accuracy = accuracy_score(y_test, dt_pred)
conf = confusion_matrix(y_test, dt_pred)
print(f'Accuracy: {accuracy}')
print(f'Confusion matrix: \n{conf}')
```
