```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

lst_profiles, lst_names ,lst_addresses, lst_emails, lst_phones = [], [], [], [], []

for i in range(0, len(links)):
    link = links[i]
    response = requests.get(link, headers = headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    tr_data = soup.find_all('tr')
    
    for row in tr_data[1:]:
        td_list = row.find_all('td')
        link_profile = base_domain + td_list[0].find_all('a')[0].get('href')
        lst_profiles.append(link_profile)
        
        names = td_list[1].text.split('\n')
        names.remove('')
        name = [i.strip() for i in names]
        name = ' '.join([i for i in name])
        lst_names.append(name)
        
        address = td_list[2].text.strip().replace(',', '-').replace('\n', '')
        lst_addresses.append(address)
        try:
            email = td_list[3].find_all('a')[0].text.strip()
        except:
            email = ''
            
        lst_emails.append(email)
        
        contact = td_list[3].text.strip()
        try:
            part1 = r'\([\d]{3}\)\s+[0-9]{3}\-[0-9]{4}'
            phone = re.search(part1, contact).group(0)
        except:
            phone = ''
        lst_phones.append(phone)
    
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(zip(lst_names, lst_addresses, lst_emails, lst_phones, lst_profiles)),
                 columns = ['Name','Address', 'Email', 'Phone', 'Link_profile'])
```
